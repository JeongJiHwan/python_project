```python
class Set:
    def __init__(self, ls=[]):
        self.ls = []
        for i in ls:
            if i not in self.ls:
                self.ls.append(i)
                
    def add(self, elem):
        if elem not in self.ls:
            self.ls.append(elem)
            
    def discard(self, elem):
        if elem in self.ls:
            self.ls.remove(elem)
            
    def clear(self):
        self.ls.clear()
        
    def __len__(self):
        return len(self.ls)
    
    def __str__(self):
        tmplist = []
        for i in self.ls:
            tmplist.append(str(i))
        return '{'+', '.join(tmplist)+'}'
    
    def __contains__(self, elem):
        if elem in self.ls:
            return True
        else:
            return False
        
    def __le__(self, other):
        for i in self.ls:
            if i not in other.ls:
                return False
        return True
    
    def __ge__(self, other):
        for i in other.ls:
            if i not in self.ls:
                return False
        return True
    
    def __or__(self, other):
        tmp = Set(self.ls)
        
        for i in other.ls:
            if i not in self.ls:
                tmp.add(i)
                
        return tmp
    
    def __and__(self, other):
        tmp = Set()
        
        for i in self.ls:
            if i in other.ls:
                tmp.add(i)
                
        return tmp
    
    def __sub__(self, other):
        tmp = Set()
        
        for i in self.ls:
            if i not in other.ls:
                tmp.add(i)
        
        return tmp
    
    def __ior__(self, other):
        for i in other.ls:
            if i not in self.ls:
                self.add(i)
        return self
    
    def __iand__(self, other):
        for i in self.ls:
            if i not in other.ls:
                self.discard(i)
        return self
    
    def __isub__(self, other):
        for i in self.ls:
            if i in other.ls:
                self.discard(i)
        return self
```

```python
a = Set([1,2,3,4,1,2,3,4])
b = Set([2,3,4,2,3,4])

print(a)
print(b)
print('-'*10)

a.discard(4)
b.discard(2)
print(a)
print(b)
print('-'*10)

print(len(a))
print(1 in a)
print(1 in b)
print('-'*10)

print(a | b)
print(a & b)
print(a - b)
print('-'*10)

print(a <= b)
print(a <= a | b)
print(a >= b)
print(a >= a & b)
print('-'*10)


a &= b
# a -= b
# a |= b
print(a)

b.clear()
print(b)
```
