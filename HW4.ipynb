### 실습2)

```python
import os

path = input('Enter a file name(*.txt) : ')
#path = 'test.txt'

if(os.path.exists(path)):
    fr = open(path, 'r')
    
    for line in fr:
        print(line.upper())
    
    fr.close()
else:
    print('%s does not exist' %path)
```
